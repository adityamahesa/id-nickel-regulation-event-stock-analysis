```python
from IPython.core.interactiveshell import InteractiveShell
from IPython.display import Markdown as md
from ipynb.fs.full import functions
import pandas as pd

InteractiveShell.ast_node_interactivity = "all"

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
from os import listdir, path
```

```python
def parse_sector(excel_filepath: str) -> str:
    excel_filepath_splitted = excel_filepath.split(sep='/')
    excel_filename = excel_filepath_splitted[-1]
    excel_filename_splitted = excel_filename.split(sep='.')
    filename = excel_filename_splitted[0]
    filename_splitted = filename.split(sep='-')
    sector = '-'.join(filename_splitted[1:])
    return sector

```

```python
__ticker_column_name = 'ticker'
__name_column_name = 'name'
__date_ipo_column_name = 'date_ipo'
__shares_column_name = 'shares'
__sector_column_name = 'sector'

ipo_bound_date = '2019-01-01'

def parse_excel(excel_filepath: str) -> pd.DataFrame:
    ticker_excel_df = pd.read_excel(io=excel_filepath, sheet_name=0)
    ticker_excel_df[__shares_column_name] = ticker_excel_df.copy()[__shares_column_name].str.replace(pat=r'\.', repl='', regex=True)
    ticker_excel_df[__shares_column_name] = ticker_excel_df.copy()[__shares_column_name].astype('int')
    ticker_excel_df = ticker_excel_df.copy()[ticker_excel_df[__date_ipo_column_name] < ipo_bound_date]
    ticker_excel_df = ticker_excel_df.copy()[[__ticker_column_name, __name_column_name, __shares_column_name]]
    ticker_excel_df[__sector_column_name] = parse_sector(excel_filepath=excel_filepath)

    return ticker_excel_df
```

```python
list_bei_path = "list-bei"

bei_excel_list = listdir(list_bei_path)
bei_excel_list = sorted(bei_excel_list)

list_df = []
for excel_filename in bei_excel_list:
    excel_filepath = path.join(list_bei_path, excel_filename)
    parsed_df = parse_excel(excel_filepath=excel_filepath)
    list_df.append(parsed_df)

tickers = pd.concat(list_df, axis=0)
tickers = tickers.copy().reset_index(drop=True)

tickers
```

```python
__tickers_filename = 'tickers.xlsx'
__tickers_sheet_name = 'tickers'

with pd.ExcelWriter(__tickers_filename) as excel_writer:
    tickers.to_excel(excel_writer=excel_writer, sheet_name=__tickers_sheet_name, index=False)
```
