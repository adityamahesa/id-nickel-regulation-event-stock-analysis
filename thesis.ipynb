```python
import util
import pandas

pandas.set_option('display.max_rows', None)
```

```python
event_date = '2019-08-30'
```

```python
tickers_top_10_sectoral_list = util.prepare_top_10_tickers_per_sector(date=event_date)
tickers_top_10_sectoral_list
```

```python
selected_tickers = tickers_top_10_sectoral_list[tickers_top_10_sectoral_list['market_value'] > 0]
selected_tickers.reset_index(drop=True, inplace=True)
selected_tickers
```

```python
start_date = "2019-06-07"
end_date = "2019-08-29"
```

```python
pandas.options.display.float_format = '{:.4%}'.format
```

```python
average_return_per_sector = util.generate_average_return_per_sector(data=selected_tickers, start_date=start_date, end_date=end_date)
average_return_per_sector
```

```python
pandas.options.display.float_format = None
```

```python
correlations = average_return_per_sector.corr()
correlations
```

```python
desired_sector = 'basic-materials'
correlation_for_desired_sector = correlations[desired_sector].to_frame()

correlation_column_name = "correlation"
first_column_name = correlation_for_desired_sector.columns[0]
correlation_for_desired_sector.rename(columns={first_column_name: correlation_column_name}, inplace=True)
correlation_for_desired_sector = correlation_for_desired_sector.sort_values(by=correlation_column_name, ascending=False)
correlation_for_desired_sector
```

```python
relativity_map = util.set_relativity_label(data=correlation_for_desired_sector)
relativity_map
```

```python
tickers = util.set_tickers_relativity_label(data=selected_tickers, relativity_map=relativity_map)
tickers
```

```python
print('number of more related tickers:', len(tickers[tickers['is_more_related'] == True]))
print('number of less related tickers:', len(tickers[tickers['is_more_related'] == False]))
```
