```python
import util
import pandas

pandas.set_option('display.max_rows', None)
```

```python
tickers_top_10_sectoral_list = util.prepare_top_10_tickers_per_sector('2019-08-30')
tickers_top_10_sectoral_list
```

```python
selected_tickers = tickers_top_10_sectoral_list[tickers_top_10_sectoral_list['market_value'] > 0]
selected_tickers
```

```python
start_date = "2019-06-07"
end_date = "2019-08-29"
```

```python
pandas.options.display.float_format = '{:.4%}'.format
average_return_per_sector = util.generate_average_return_per_sector(selected_tickers, start_date=start_date, end_date=end_date)
average_return_per_sector
pandas.options.display.float_format = None
```

```python
correlations = average_return_per_sector.corr()
correlations
```

```python
desired_sector = 'basic-materials'
correlation_for_desired_sector = correlations[desired_sector].to_frame()
first_column_name = correlation_for_desired_sector.columns[0]
correlation_for_desired_sector = correlation_for_desired_sector.sort_values(first_column_name, ascending=False)
correlation_for_desired_sector
```
